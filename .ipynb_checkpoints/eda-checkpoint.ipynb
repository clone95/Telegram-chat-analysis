```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
import seaborn as sns
import json 
import os
import sys  
from ipywidgets import interact, interactive, fixed, interact_manual
from IPython.display import display
import ipywidgets as widgets

matplotlib.rcParams['figure.figsize'] = (15.0, 7.0)
matplotlib.rcParams['font.size']=20
dataset_names = os.listdir('data')

data_sel = widgets.Dropdown(
    options=[x.split('.')[0] for x in dataset_names],
    description='Dataset:',
    disabled=False,
)
resample_slider = widgets.SelectionSlider(
    options=['m', 'w', 'd'],
    description='Resample:',
    disabled=False,
    continuous_update=True,
    orientation='horizontal',
    readout=True
)

display(data_sel)
```

```python
data = pd.read_csv('data.csv')
data['date'] = data.date.apply(lambda x: pd.to_datetime(x))
data['month'] = data.date.apply(lambda x: x.month)
data['day'] = data.date.apply(lambda x: x.day)
data['hour'] = data.date.apply(lambda x: x.hour)
data['minute'] = data.date.apply(lambda x: x.minute)
data['second'] = data.date.apply(lambda x: x.second)
data = data.set_index('date')
data = data.drop(['Unnamed: 0'], axis=1)
data.head()
```

```python
grouped_by_from = data.groupby('da').count()
ax = sns.barplot(x=grouped_by_from.index, y="text", data=grouped_by_from).set_title('Quanti messaggi in totale sono stati inviati?', fontsize=20);
```

```python
monthly = data[['text', 'month']].groupby('month').count()
monthly.index = ['Aprile', 'Maggio', 'Giugno', 'Luglio']
monthly.plot(linewidth=4, title='Numero di messaggi inviati rispetto al mese');
```

### Choose _resample rate_

```python
display(resample_slider)
```

```python
data[['text']].resample(resample_slider.value).count().plot(linewidth=4, title='Numero di messaggi inviati rispetto per ora del giorno');
```
